(1) Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`. After generating the list, find the following:


(i) Write a Python function to calculate the mean of a given list of numbers.
 Create a function to find the median of a list of numbers.

(ii) Develop a program to compute the mode of a list of integers.

(iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

(iv) Write a Python function to find the geometric mean of a list of positive numbers.

(v) Create a program to calculate the harmonic mean of a list of values.

(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.


````python
import numpy as np
from collections import Counter

# Generate a list of 100 integers between 90 and 130
np.random.seed(0)  # For reproducibility
int_list = np.random.randint(90, 131, size=100).tolist()

# Function definitions

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]

def calculate_mode(numbers):
    count = Counter(numbers)
    max_count = max(count.values())
    return [num for num, cnt in count.items() if cnt == max_count]

def calculate_weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

def calculate_geometric_mean(numbers):
    return np.exp(np.mean(np.log(numbers)))

def calculate_harmonic_mean(numbers):
    n = len(numbers)
    return n / sum(1 / x for x in numbers)

def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

def calculate_trimmed_mean(numbers, percentage):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    trim_count = int(n * percentage / 100)
    trimmed_numbers = sorted_numbers[trim_count:n - trim_count]
    return calculate_mean(trimmed_numbers)

# Example usage
print("List of integers:", int_list)
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a list of 500 integers between 200 and 300
np.random.seed(0)
int_list2 = np.random.randint(200, 301, size=500).tolist()

# Visualization
data = np.array(int_list2)

plt.figure(figsize=(14, 6))

# Frequency Distribution
plt.subplot(1, 3, 1)
sns.histplot(data, kde=False, bins=30)
plt.title('Frequency Distribution')

# Smoothened KDE Plot
plt.subplot(1, 3, 2)
sns.kdeplot(data, bw_adjust=0.5)
plt.title('Smoothened KDE Plot')

# Gaussian Distribution & KDE Plot
plt.subplot(1, 3, 3)
sns.histplot(data, kde=True, bins=30)
plt.title('Gaussian & KDE Plot')

plt.tight_layout()
plt.show()

# Functions

def calculate_range(numbers):
    return max(numbers) - min(numbers)

def calculate_variance(numbers):
    mean = np.mean(numbers)
    return np.mean((np.array(numbers) - mean) ** 2)

def calculate_standard_deviation(numbers):
    return np.sqrt(calculate_variance(numbers))

def calculate_iqr(numbers):
    q75, q25 = np.percentile(numbers, [75 ,25])
    return q75 - q25

def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers)
    return std_dev / mean

def calculate_mad(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(np.array(numbers) - mean))

def calculate_quartile_deviation(numbers):
    q75, q25 = np.percentile(numbers, [75 ,25])
    return (q75 - q25) / 2

def calculate_range_based_coefficient_of_dispersion(numbers):
    range_val = calculate_range(numbers)
    mean = np.mean(numbers)
    return range_val / mean

# Example Usage
print("Range:", calculate_range(int_list2))
print("Variance:", calculate_variance(int_list2))
print("Standard Deviation:", calculate_standard_deviation(int_list2))
print("IQR:", calculate_iqr(int_list2))
print("Coefficient of Variation:", calculate_coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD):", calculate_mad(int_list2))
print("Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("Range-Based Coefficient of Dispersion:", calculate_range_based_coefficient_of_dispersion(int_list2))
````

 (3) Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, probabilities, values):
        """
        Initializes the DiscreteRandomVariable with given probabilities and corresponding values.

        :param probabilities: List of probabilities associated with each value.
        :param values: List of values of the random variable.
        """
        if len(probabilities) != len(values):
            raise ValueError("Length of probabilities and values must be the same.")
        if not abs(sum(probabilities) - 1.0) < 1e-9:
            raise ValueError("Probabilities must sum up to 1.")
        self.probabilities = probabilities
        self.values = values

    def expected_value(self):
        """
        Calculates the expected value (mean) of the discrete random variable.

        :return: The expected value.
        """
        return sum(p * v for p, v in zip(self.probabilities, self.values))

    def variance(self):
        """
        Calculates the variance of the discrete random variable.

        :return: The variance.
        """
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for p, v in zip(self.probabilities, self.values))

# Example usage
if __name__ == "__main__":
    values = [1, 2, 3, 4]
    probabilities = [0.1, 0.2, 0.3, 0.4]

    # Create an instance of DiscreteRandomVariable
    dr_variable = DiscreteRandomVariable(probabilities, values)

    # Calculate and print the expected value and variance
    print("Expected Value:", dr_variable.expected_value())
    print("Variance:", dr_variable.variance())
````

(4) Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.


````python
import random

def simulate_die_rolls(num_rolls):
    """
    Simulates rolling a fair six-sided die num_rolls times.

    :param num_rolls: Number of times to roll the die.
    :return: List of outcomes of the rolls.
    """
    return [random.randint(1, 6) for _ in range(num_rolls)]

def calculate_expected_value(outcomes):
    """
    Calculates the expected value of the outcomes.

    :param outcomes: List of outcomes from the die rolls.
    :return: Expected value.
    """
    return sum(outcomes) / len(outcomes)

def calculate_variance(outcomes, expected_value):
    """
    Calculates the variance of the outcomes.

    :param outcomes: List of outcomes from the die rolls.
    :param expected_value: Expected value of the outcomes.
    :return: Variance.
    """
    return sum((x - expected_value) ** 2 for x in outcomes) / len(outcomes)

# Simulate rolling the die 10000 times
num_rolls = 10000
rolls = simulate_die_rolls(num_rolls)

# Calculate expected value and variance
expected_value = calculate_expected_value(rolls)
variance = calculate_variance(rolls, expected_value)

print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")

# Theoretical values for a fair six-sided die
theoretical_expected_value = 3.5
theoretical_variance = (1**2 + 2**2 + 3**2 + 4**2 + 5**2 + 6**2) / 6 - theoretical_expected_value**2

print(f"Theoretical Expected Value: {theoretical_expected_value:.2f}")
print(f"Theoretical Variance: {theoretical_variance:.2f}")
````

(5) Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.


````python
import numpy as np

def generate_samples(distribution, params, num_samples):
    """
    Generates random samples from a specified probability distribution and calculates their mean and variance.

    :param distribution: The type of probability distribution ('binomial' or 'poisson').
    :param params: Parameters for the distribution.
                   For 'binomial': (n, p)
                   For 'poisson': (lambda)
    :param num_samples: Number of random samples to generate.
    :return: Tuple of (mean, variance) of the generated samples.
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, num_samples)
    elif distribution == 'poisson':
        lambda_ = params
        samples = np.random.poisson(lambda_, num_samples)
    else:
        raise ValueError("Unsupported distribution type. Use 'binomial' or 'poisson'.")

    # Calculate mean and variance
    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance

# Example usage
if __name__ == "__main__":
    # Generate samples from a binomial distribution
    binomial_params = (10, 0.5)  # n=10, p=0.5
    num_samples = 10000
    binomial_mean, binomial_variance = generate_samples('binomial', binomial_params, num_samples)
    print(f"Binomial Distribution - Mean: {binomial_mean:.2f}, Variance: {binomial_variance:.2f}")

    # Generate samples from a Poisson distribution
    poisson_params = 5  # lambda=5
    poisson_mean, poisson_variance = generate_samples('poisson', poisson_params, num_samples)
    print(f"Poisson Distribution - Mean: {poisson_mean:.2f}, Variance: {poisson_variance:.2f}")
````

(6) Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.


````python
import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
    """
    Generates random samples from a Gaussian (normal) distribution and calculates their mean, variance, and standard deviation.

    :param mean: Mean of the Gaussian distribution.
    :param std_dev: Standard deviation of the Gaussian distribution.
    :param num_samples: Number of random samples to generate.
    :return: Tuple of (mean, variance, standard deviation) of the generated samples.
    """
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(mean, std_dev, num_samples)

    # Calculate mean, variance, and standard deviation
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.sqrt(sample_variance)

    return sample_mean, sample_variance, sample_std_dev

# Example usage
if __name__ == "__main__":
    # Parameters for the Gaussian distribution
    mean = 0          # Mean of the distribution
    std_dev = 1       # Standard deviation of the distribution
    num_samples = 10000  # Number of samples to generate

    # Generate samples and calculate statistics
    sample_mean, sample_variance, sample_std_dev = generate_gaussian_samples(mean, std_dev, num_samples)

    # Print the results
    print(f"Mean of the samples: {sample_mean:.2f}")
    print(f"Variance of the samples: {sample_variance:.2f}")
    print(f"Standard Deviation of the samples: {sample_std_dev:.2f}")
````

````python
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.stats import skew
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Extract columns for analysis
total_bill = tips['total_bill']
tip = tips['tip']

# (i) Calculate skewness
def calculate_skewness(data):
    """
    Calculates the skewness of a given dataset.

    :param data: List or Pandas Series of numerical data.
    :return: Skewness of the data.
    """
    return skew(data, nan_policy='omit')

# Calculate skewness for total_bill and tip
total_bill_skewness = calculate_skewness(total_bill)
tip_skewness = calculate_skewness(tip)

print(f"Skewness of total_bill: {total_bill_skewness:.2f}")
print(f"Skewness of tip: {tip_skewness:.2f}")

# (ii) Determine skewness type
def skewness_type(skewness):
    """
    Determines the type of skewness.

    :param skewness: Skewness value.
    :return: Type of skewness ('positive', 'negative', 'symmetric').
    """
    if skewness > 0:
        return 'positive'
    elif skewness < 0:
        return 'negative'
    else:
        return 'symmetric'

print(f"total_bill skewness type: {skewness_type(total_bill_skewness)}")
print(f"tip skewness type: {skewness_type(tip_skewness)}")

# (iii) Calculate covariance
def calculate_covariance(x, y):
    """
    Calculates the covariance between two datasets.

    :param x: First dataset.
    :param y: Second dataset.
    :return: Covariance between x and y.
    """
    return np.cov(x, y, ddof=0)[0, 1]

covariance = calculate_covariance(total_bill, tip)
print(f"Covariance between total_bill and tip: {covariance:.2f}")

# (iv) Calculate Pearson correlation coefficient
def calculate_pearson_correlation(x, y):
    """
    Calculates the Pearson correlation coefficient between two datasets.

    :param x: First dataset.
    :param y: Second dataset.
    :return: Pearson correlation coefficient between x and y.
    """
    return np.corrcoef(x, y)[0, 1]

pearson_correlation = calculate_pearson_correlation(total_bill, tip)
print(f"Pearson correlation coefficient between total_bill and tip: {pearson_correlation:.2f}")

# (v) Visualize the correlation using scatter plot
def plot_scatter(df, x_col, y_col):
    """
    Plots a scatter plot to visualize the correlation between two columns in a DataFrame.

    :param df: Pandas DataFrame containing the data.
    :param x_col: Name of the column for x-axis.
    :param y_col: Name of the column for y-axis.
    """
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=x_col, y=y_col)
    plt.title(f"Scatter Plot of {x_col} vs {y_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()

# Plot scatter plot for total_bill and tip
plot_scatter(tips, 'total_bill', 'tip')
````

(8) Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

````python
import numpy as np

def normal_pdf(x, mean, std_dev):
    """
    Calculates the Probability Density Function (PDF) of a normal distribution.

    :param x: The value at which to evaluate the PDF.
    :param mean: Mean of the normal distribution.
    :param std_dev: Standard deviation of the normal distribution.
    :return: PDF value at x.
    """
    # Calculate the PDF using the normal distribution formula
    coefficient = 1 / (std_dev * np.sqrt(2 * np.pi))
    exponent = np.exp(- (x - mean)**2 / (2 * std_dev**2))
    pdf_value = coefficient * exponent
    
    return pdf_value

# Example usage
if __name__ == "__main__":
    mean = 0       # Mean of the normal distribution
    std_dev = 1    # Standard deviation of the normal distribution
    x = 1.0        # Value at which to calculate the PDF

    pdf_value = normal_pdf(x, mean, std_dev)
    print(f"PDF value at x = {x}: {pdf_value:.4f}")
````

(9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python
import numpy as np

def exponential_cdf(x, rate):
    """
    Calculates the Cumulative Distribution Function (CDF) of an exponential distribution.

    :param x: The value at which to evaluate the CDF (must be non-negative).
    :param rate: The rate parameter (lambda) of the exponential distribution.
    :return: CDF value at x.
    """
    if x < 0:
        raise ValueError("x must be non-negative.")
    
    # Calculate the CDF using the exponential distribution formula
    cdf_value = 1 - np.exp(-rate * x)
    
    return cdf_value

# Example usage
if __name__ == "__main__":
    rate = 1.5   # Rate parameter (lambda) of the exponential distribution
    x = 2.0      # Value at which to calculate the CDF

    cdf_value = exponential_cdf(x, rate)
    print(f"CDF value at x = {x} with rate = {rate}: {cdf_value:.4f}")
````

(10)Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
import numpy as np
from math import factorial

def poisson_pmf(k, lambda_):
    """
    Calculates the Probability Mass Function (PMF) of a Poisson distribution.

    :param k: The number of occurrences (must be a non-negative integer).
    :param lambda_: The rate parameter (lambda) of the Poisson distribution.
    :return: PMF value at k.
    """
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    
    # Calculate the PMF using the Poisson distribution formula
    pmf_value = (lambda_**k * np.exp(-lambda_)) / factorial(k)
    
    return pmf_value

# Example usage
if __name__ == "__main__":
    lambda_ = 3.0  # Rate parameter (lambda) of the Poisson distribution
    k = 4         # Number of occurrences

    pmf_value = poisson_pmf(k, lambda_)
    print(f"PMF value for k = {k} with lambda = {lambda_}: {pmf_value:.4f}")
````

(11) A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.

To generate the data use the following command:
 ```python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
 # 70 purchases out of 1000 visitors new_layout = np.array([1] * 70 + [0] * 930)
```
Apply z-test to find which layout is successful.

````python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Count the number of successes and trials
x_old = np.sum(old_layout)  # Number of conversions for the old layout
n_old = len(old_layout)     # Number of visitors for the old layout
x_new = np.sum(new_layout)  # Number of conversions for the new layout
n_new = len(new_layout)     # Number of visitors for the new layout

def compare_conversion_rates(x_old, n_old, x_new, n_new):
    """
    Compare conversion rates between two website layouts using a two-proportion z-test.

    :param x_old: Number of conversions for the old layout.
    :param n_old: Number of visitors to the old layout.
    :param x_new: Number of conversions for the new layout.
    :param n_new: Number of visitors to the new layout.
    :return: z-statistic and p-value for the test.
    """
    # Convert counts to arrays for the z-test function
    counts = np.array([x_old, x_new])
    nobs = np.array([n_old, n_new])
    
    # Perform the z-test
    z_stat, p_value = proportions_ztest(counts, nobs, alternative='larger')
    
    return z_stat, p_value

# Perform the comparison
z_stat, p_value = compare_conversion_rates(x_old, n_old, x_new, n_new)

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout has a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

(14) A customer service department claims that their average response time is less than 5 minutes. A sample of recent customer interactions was taken, and the response times were recorded.
Implement the below code to generate the array of response time:
 ```python
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])
 ```
Implement z-test to find the claims made by customer service department are tru or false.

````python
from scipy import stats
import numpy as np

# Generate the data
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

def one_sample_z_test(sample, population_mean):
    """
    Perform a one-sample z-test to determine if the sample mean is significantly different from the population mean.

    :param sample: Array of sample data.
    :param population_mean: Hypothesized population mean.
    :return: z-statistic and p-value for the test.
    """
    # Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=0)  # Use population standard deviation (ddof=0)
    n = len(sample)
    
    # Calculate the z-statistic
    z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
    
    # Calculate the p-value for a one-tailed test
    p_value = stats.norm.cdf(z_stat)
    
    return z_stat, p_value

# Define the population mean (claimed average response time)
population_mean = 5

# Perform the z-test
z_stat, p_value = one_sample_z_test(response_times, population_mean)

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average response time is significantly less than 5 minutes.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the average response time is less than 5 minutes.")
````

````python
from scipy import stats
import numpy as np

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    """
    Perform an A/B test analysis to compare the means of two independent samples.

    :param layout_a_clicks: List of click-through rates for layout A.
    :param layout_b_clicks: List of click-through rates for layout B.
    :return: t-statistic, degrees of freedom, and p-value for the test.
    """
    # Convert lists to numpy arrays
    layout_a = np.array(layout_a_clicks)
    layout_b = np.array(layout_b_clicks)
    
    # Calculate means
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    
    # Calculate variances
    var_a = np.var(layout_a, ddof=1)  # Sample variance
    var_b = np.var(layout_b, ddof=1)  # Sample variance
    
    # Calculate sample sizes
    n_a = len(layout_a)
    n_b = len(layout_b)
    
    # Calculate pooled standard deviation
    pooled_std = np.sqrt(((n_a - 1) * var_a + (n_b - 1) * var_b) / (n_a + n_b - 2))
    
    # Calculate t-statistic
    t_stat = (mean_a - mean_b) / (pooled_std * np.sqrt(1 / n_a + 1 / n_b))
    
    # Calculate degrees of freedom
    df = n_a + n_b - 2
    
    # Calculate p-value (two-tailed test)
    p_value = 2 * stats.t.cdf(-abs(t_stat), df)
    
    return t_stat, df, p_value

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test analysis
t_stat, df, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
else:
    print("Fail to reject the null hypothesis: No significant difference in click-through rates between the two layouts.")
````

````python
from scipy import stats
import numpy as np

def two_sample_t_test(sample1, sample2):
    """
    Perform a two-sample t-test to compare the means of two independent samples.

    :param sample1: Array of cholesterol levels for the existing drug.
    :param sample2: Array of cholesterol levels for the new drug.
    :return: t-statistic, degrees of freedom, and p-value for the test.
    """
    # Convert lists to numpy arrays
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)
    
    # Calculate means
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    
    # Calculate variances
    var1 = np.var(sample1, ddof=1)  # Sample variance
    var2 = np.var(sample2, ddof=1)  # Sample variance
    
    # Calculate sample sizes
    n1 = len(sample1)
    n2 = len(sample2)
    
    # Calculate the t-statistic
    t_stat, p_value = stats.ttest_ind(sample1, sample2, alternative='less')  # One-tailed test
    
    # Calculate degrees of freedom
    # For Welch’s t-test
    df = (var1 / n1 + var2 / n2) ** 2 / ((var1 / n1) ** 2 / (n1 - 1) + (var2 / n2) ** 2 / (n2 - 1))
    
    return t_stat, df, p_value

# Data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform the t-test
t_stat, df, p_value = two_sample_t_test(existing_drug_levels, new_drug_levels)

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new drug is significantly more effective in reducing cholesterol levels than the existing drug.")
else:
    print("Fail to reject the null hypothesis: No significant difference in cholesterol levels between the new drug and the existing drug.")
````
